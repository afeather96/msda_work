```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_csv('teleco_market_basket.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df['nulls_sum'] = df.isnull().sum(axis=1)
df.head()
```

```python
empty_rows = np.where(df['nulls_sum']==20)[0]
empty_rows
```

```python
len(empty_rows)
```

```python
df_clean = df.copy()
```

```python
df_clean = df_clean.drop(list(empty_rows), axis=0)
df_clean.head()
```

```python
df_clean.shape
```

```python
df_clean = df_clean.drop('nulls_sum', axis=1)
```

```python
df_clean.reset_index(inplace=True)
df_clean.head()
```

```python
df_clean.drop('index', axis=1, inplace=True)
```

```python
df_clean.shape
```

```python
transactions = []

for i in range(df_clean.shape[0]):
    transactions.append([str(df_clean.values[i, j]) for j in range(df_clean.shape[1])])
    
transactions
```

```python
len(transactions)
```

```python
df_clean.dtypes
```

```python
te = TransactionEncoder()

clean_transactions = te.fit(transactions).transform(transactions)
```

```python
print(te.columns_)
```

```python
final_df = pd.DataFrame(clean_transactions, columns=te.columns_)
final_df.head()
```

```python
final_df.shape
```

```python
final_df.dtypes
```

```python
final_df.isnull().sum().sum()
```

```python
for col in final_df.columns:
    print(col)
```

```python
final_df = final_df.drop('nan', axis=1)
print(final_df.shape)
```

```python
apriori_rules = apriori(final_df, min_support=0.02, use_colnames=True)
apriori_rules.head()
```

```python
print(apriori_rules.shape)
```

```python
final_rules = association_rules(apriori_rules, metric='lift', min_threshold=1)
final_rules.head()
```

```python
print(final_rules.shape)
```

```python
filtered_rules = final_rules[final_rules['confidence'] >= 0.35]
filtered_rules.head()
```

```python
print(len(filtered_rules))
```

```python
sorted_rules = filtered_rules.sort_values(by='lift', ascending=False).head(3)
sorted_rules
```
